```python
from skimage import data,io,filters
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage.transform import resize
from skimage.util import random_noise
from skimage.filters.rank import mean
from skimage.morphology import square
from skimage.filters.rank import median
from skimage.filters import gaussian

import numpy as np
```

a) Import and resize to 512 x 512  an image of your face. Then convert it to greyscale and display the minimum, maximum and mean grayscale values

```python
#Import the image
image = io.imread("face.jpg")
#Resize image
image = resize(image, (512, 512))
#Greyscale it
grey_image = rgb2gray(image)
```

```python
print("The greyscale minimum value is", grey_image.min(), "The greyscale maximum value is ", grey_image.max(), "The greyscale mean value is ", grey_image.mean())
```

```python
fig, ax = plt.subplots(1, 2 )
ax[0] = plt.subplot(1, 2, 1)
ax[1] = plt.subplot(1, 2, 2)

ax[0].imshow(image, cmap = plt.cm.gray)
ax[1].imshow(grey_image, cmap = plt.cm.gray)
```

b) Taking the output image from part (a) and add Gaussian noise (mean=0.0, variance=0.01). Display your result.

```python
noised_image = random_noise(grey_image, mode='gaussian', mean=0.0, var=0.01)
```

```python
fig, ax = plt.subplots(1, 2 )
ax[0] = plt.subplot(1, 2, 1)
ax[1] = plt.subplot(1, 2, 2)

ax[0].imshow(grey_image, cmap = plt.cm.gray)
ax[1].imshow(noised_image, cmap = plt.cm.gray)
```

```python
mean = mean(noised_image, square(5))
```

```python
filtered = median(noised_image, square(5))
```

```python
if mean.all() == noised_image.all():
  print("similar")
```

```python
fig, ax = plt.subplots(1, 3 )
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)
ax[0].imshow(noised_image, cmap = plt.cm.gray)
ax[1].imshow(mean, cmap = plt.cm.gray)
ax[2].imshow(filtered, cmap = plt.cm.gray)
```

```python
gauss_img = gaussian(noised_image, sigma = 1)
gauss_img2 = gaussian(noised_image, sigma = 0.5)
gauss_img3 = gaussian(noised_image, sigma = 2)

fig, ax = plt.subplots(1, 3 )
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)
ax[0].imshow(gauss_img, cmap = plt.cm.gray)
ax[1].imshow(gauss_img2, cmap = plt.cm.gray)
ax[2].imshow(gauss_img3, cmap = plt.cm.gray)
```
