```python
from skimage import data,io,filters
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage.transform import resize
```

a) Import and resize to 512 x 512  an image of your face. Then convert it to greyscale and display the minimum, maximum and mean grayscale values

```python
#Import the image
image = io.imread("face.jpg")
#Resize image
image = resize(image, (512, 512))
#Greyscale it
grey_image = rgb2gray(image)
```

```python
print("The greyscale minimum value is", grey_image.min(), "The greyscale maximum value is ", grey_image.max(), "The greyscale mean value is ", grey_image.mean())
```

```python
fig, ax = plt.subplots(1, 2 )
ax[0] = plt.subplot(1, 2, 1)
ax[1] = plt.subplot(1, 2, 2)

ax[0].imshow(image, cmap = plt.cm.gray)
ax[1].imshow(grey_image, cmap = plt.cm.gray)
```
